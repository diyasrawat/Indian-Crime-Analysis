```python
# Data for different crime categories
data = {
    'STATE NAME': ['KERALA', 'MADHYA PRADESH', 'MAHARASHTRA', 'RAJASTHAN', 'UTTAR PRADESH', 'NCT OF DELHI' ],
    'KIDNAPPED & ABDUCTED (TOTAL)': [367, 16109, 13255, 10245, 20329,15153],
    'TOTAL MOTIVES': [305, 2004, 2299, 1551, 4889,528],
    'MURDERED TOTAL (TOTAL)': [431, 2068, 2395, 1571, 5055,545],
    'RAPE VICTIMS (TOTAL)': [1661, 4908, 4216, 3657, 4817, 2170],
    'TRAFFICED (TOTAL)': [614, 4817, 1141, 5767, 924, 264 ],
    'VIOLENT TOTAL': [13548, 26719, 42468, 16223, 65090, 16648],
}


# Create DataFrame
crime_df = pd.DataFrame(data)

# Display the resulting DataFrame
print(crime_df)
```

```python
# Add a column for the sum of crime scores for each state
crime_df['Total Crime Cases'] = crime_df.iloc[:, 1:].sum(axis=1)

# Display the resulting DataFrame
print(crime_df[['STATE NAME', 'Total Crime Cases']])
```

```python
uttar_pradesh_scores = crime_df[crime_df['STATE NAME'] == 'UTTAR PRADESH']
print(uttar_pradesh_scores)
```

```python
crime_categories = ['KIDNAPPED & ABDUCTED (TOTAL)', 'TOTAL MOTIVES', 'MURDERED TOTAL (TOTAL)',
                    'RAPE VICTIMS (TOTAL)', 'TRAFFICED (TOTAL)', 'VIOLENT TOTAL']

# Total crime score for Uttar Pradesh
total_crime_score_up = 101104

# Calculate the percentage for each crime category based on the total crime score
for category in crime_categories:
    uttar_pradesh_scores[category + ' (%)'] = (uttar_pradesh_scores[category] / total_crime_score_up) * 100

# Print the percentages for each crime category in Uttar Pradesh
for category in crime_categories:
    print(f'{category} Percentage: {uttar_pradesh_scores.iloc[0][category + " (%)"]:.2f}%')
```

```python
import matplotlib.pyplot as plt

# Assuming uttar_pradesh_scores is your DataFrame with the calculated percentages

# Set the crime categories as x-axis labels
categories = [category + ' (%)' for category in crime_categories]

# Plot the percentages
plt.figure(figsize=(10, 6))
plt.bar(categories, uttar_pradesh_scores.iloc[0][categories])
plt.title('Percentage of Crime Categories in Uttar Pradesh')
plt.xlabel('Crime Categories')
plt.ylabel('Percentage')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming uttar_pradesh_scores is your DataFrame with the calculated percentages

# Set the crime categories and their corresponding percentages
categories = [category + ' (%)' for category in crime_categories]
percentages = uttar_pradesh_scores.iloc[0][categories]

# Plot a donut pie chart
fig, ax = plt.subplots(figsize=(8, 8))

# Draw the pie chart
wedges, texts, autotexts = ax.pie(percentages, labels=categories, autopct='%1.1f%%', startangle=90,
                                  wedgeprops=dict(width=0.3, edgecolor='w'))

# Add a circle in the center to create a donut chart
centre_circle = plt.Circle((0, 0), 0.2, color='white', edgecolor='black', linewidth=0.8)
ax.add_patch(centre_circle)

# Customize the chart
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart of Crime Categories in Uttar Pradesh')

# Show the chart
plt.show()
```
