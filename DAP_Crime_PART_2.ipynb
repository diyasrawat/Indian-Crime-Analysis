```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Filter data for Uttar Pradesh
up_data = murder_motive[murder_motive['STATE NAME'] == 'UTTAR PRADESH']

up_data = up_data.drop(['Unnamed: 0', 'STATE NAME', 'TOTAL MOTIVES'], axis=1)

# Calculate the total motives for each category
up_data_total = up_data.sum(axis=0)

# Find the major reason (motive) for murder cases
major_motive = up_data_total.idxmax()

# Calculate percentages relative to the total number of motives
percentage_values = (up_data_total / up_data_total.sum()) * 100

# Plotting the data
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=percentage_values.index, y=percentage_values.values, palette='viridis')

# Adding percentage labels on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title(f"Major Reason for Murder Motive Cases in Uttar Pradesh: {major_motive}")
plt.xlabel("Motives")
plt.ylabel("Percentage of Total Motives")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

up_victims = murder_victim[murder_victim['Area_Name'] == 'Uttar Pradesh']

age_gender_data = up_victims[['Year', 'Gender', 'Victims_Upto_10_Yrs', 'Victims_Upto_15_18_Yrs', 'Victims_Upto_18_30_Yrs', 'Victims_Upto_30_50_Yrs', 'Victims_Above_50_Yrs']]

# Group by Year and Gender, and sum the victim counts
grouped_data = age_gender_data.groupby(['Year', 'Gender']).sum().reset_index()

# Display the resulting DataFrame
print(grouped_data)
```

```python
import matplotlib.pyplot as plt

# Extract age-wise victim counts across all years
age_columns = ['Victims_Upto_10_Yrs', 'Victims_Upto_15_18_Yrs', 'Victims_Upto_18_30_Yrs', 'Victims_Upto_30_50_Yrs', 'Victims_Above_50_Yrs']
age_wise_counts_all_years = grouped_data[age_columns].sum()

# Plotting the pie chart for age-wise distribution across all years
plt.figure(figsize=(7, 7))
plt.pie(age_wise_counts_all_years, labels=age_wise_counts_all_years.index, autopct='%1.1f%%', startangle=90, colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FFD700'])
plt.title('Age-wise Distribution of Murder Victims in Uttar Pradesh (All Years)')
plt.show()
```

```python
# Extract gender-wise victim counts across all years
gender_wise_counts_all_years = grouped_data.groupby('Gender')

# Plotting the bar plot for gender-wise distribution across all years
gender_wise_counts_all_years.plot(kind='bar', color=['#FF9999', '#66B2FF'])
plt.title('Gender-wise Distribution of Murder Victims (All Years)')
plt.xlabel('Gender')
plt.ylabel('Number of Victims')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Select the relevant columns for age groups
age_group_columns = [
    'RAPE VICTIMS BELOW 6 YEARS', 'RAPE VICTIMS 6-11 YEARS', 'RAPE VICTIMS 12-15 YEARS', 'RAPE VICTIMS 16-17 YEARS',
    'RAPE VICTIMS 18-29 YEARS', 'RAPE VICTIMS 30-44 YEARS', 'RAPE VICTIMS 45-59 YEARS', 'RAPE VICTIMS 60 YEARS & ABOVE'
]

# Filter data for Uttar Pradesh
uttar_pradesh_data = rape_victim[rape_victim['STATE NAME'] == 'UTTAR PRADESH']

# Select only the specified age group columns
age_group_data = uttar_pradesh_data[age_group_columns]

# Calculate the total rape victims
total_rape_victims = uttar_pradesh_data['RAPE VICTIMS (TOTAL)'].iloc[0]

# Calculate the percentage for each age group
percentages = (age_group_data.sum() / total_rape_victims) * 100

# Plot a bar plot with percentages on top of each bar
plt.figure(figsize=(10, 6))
bars = plt.bar(percentages.index, percentages, color='skyblue')

# Add percentages on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.1f}%', ha='center', va='bottom')

plt.title('Percentage of Rape Victims by Age Group in Uttar Pradesh')
plt.xlabel('Age Group')
plt.ylabel('Percentage')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=percentages.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Pie Chart: Percentage of Rape Victims by Age Group in Uttar Pradesh')
plt.show()
```
